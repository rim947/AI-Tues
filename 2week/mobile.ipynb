## csv, xls을 웹주소와 로컬 파일로 각각 읽어서 데이터프레임 작성

```python
### pandas 라이브러리
import pandas as pd

# 1-1. URL 웹주소로 데이터 불러오기
url = "https://raw.githubusercontent.com/MyungKyuYi/AI-class/main/mobile.csv" 
df = pd.read_csv(url)
df.head()
```

```python
# 1-2 파일 경로로 데이터 불러오기
import os 
file = 'mobile.csv'

file_path = os.path.join(os.getcwd(), file)

df2=pd.read_csv(file_path)

if os.path.exists(file_path):
    df2 = pd.read_csv(file_path)
    print(df2.head())
else:
    print("File not found.")
```

```python
# 1-3. 웹에서 url을 읽어서 xlsx 불러오기
url2 = "https://raw.githubusercontent.com/MyungKyuYi/AI-class/main/combined_dataset-1.xlsx"
df3 = pd.read_excel(url2)
df3.head()
```

```python
# 1-4 파일 경로로 xlsx 불러오기
file4 = 'combined_dataset-1.xlsx'

file_path4 = os.path.join(os.getcwd(), file4)

df4=pd.read_excel(file_path4)

if os.path.exists(file_path4):
    df = pd.read_excel(file_path4)
    print(df.head())
else:
    print("File not found.")
```
